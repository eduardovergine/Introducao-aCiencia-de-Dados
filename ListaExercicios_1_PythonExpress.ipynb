```python
# 3) Elabore um código que peça dois números e imprima a soma.
numero1 = input ("Digite o primeiro número: ")
numero2 = input ("Digite o segundo número: ")
soma = int(numero1) + int(numero2)
print("A soma do números é: ", soma)
```

```python
# 4) Elabore um código que peça as 4 notas bimestrais e mostre a média.
nota1 = input ("Digite a nota do primeiro bimestre ")
nota2 = input ("Digite a nota do segundo bimestre ")
nota3 = input ("Digite a nota do terceiro bimestre ")
nota4 = input ("Digite a nota do quarto bimestre ")
media = ((int(nota1) + int(nota2) + int(nota3) + int(nota4))/4)
print("Média bimestral: ", media)
```

```python
# 6) Elabore um código que peça o raio de um círculo, calcule e mostre sua área.
raio = input("Digite o raio do círculo: ")
area = 3.14 * (int(raio)**2)
print ("Área do círculo: " + str(area) + "m²")
```

```python
# 7) Elabore um código que calcule a área de um quadrado, em seguida mostre o dobro desta área para o usuário.
lado  = input("Digite o valor do lado do quadrado: ")
area = int(lado)**2
dobro_area = area*2
print(f'Área do quadrado: {area}m²\nDobro da área do quadrado: {dobro_area}m²')
```

```python
# 8) Elabore um código que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.
# Calcule e mostre o total do seu salário no referido mês em uma mensagem amigável.
salario_hora = input("Quanto você ganha por hora? ")
horas_mensais = input("Quantas horas trabalhadas no mês? ")
total_salario = int(salario_hora) * int(horas_mensais)
print(f"Este é o total do seu salário: R${total_salario}")
```

```python
# 9) Elabore um código que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius. ** Onde C = 5 * ((F-32) / 9)**.
F = input("Digite a temperatura em Fahrenheit: ")
C = round((((int(F)-32)*5) / 9), 0)
print (f'Temperatura em Celsius: {int(C)}°C')
```

```python
# 10) Elabore um código que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.
celsius = input("Digite a temperatura em Celsius: ")
fheit = round(((int(celsius) * 9 / 5) + 32), 0)
print (f"Temperatura em Fahrenheit: {int(fheit)}°F")
```
