```python
#18) Elabore uma função chamada calcula_desconto que receba dois parâmetros: o valor de um produto e a porcentagem de desconto.
#A função deve retornar o valor final do produto com o desconto aplicado. Teste a função e imprima o resultado.

def calcula_desconto(valor, porcentagem):
  return valor * (1 - porcentagem/100)

calcula_desconto(100, 10)
```

```python
#19) Elabore uma função chamada classifica_nota que receba uma nota de 0 a 10. A função deve retornar "Aprovado" se a nota for maior ou igual a 7, "Recuperação" se estiver entre 5 e 6.9, e "Reprovado" se for menor que 5.
#Chame a função passando uma nota e imprima o retorno.

def classifica_nota(nota):
  if nota >= 7:
    print("Aprovado")
  elif nota < 5:
    print("Reprovado")
  else:
    print("Recuperação")

classifica_nota(6.5)
```

```python
#20) Elabore um código que receba uma lista de números inteiros mistos (positivos, negativos e zeros).
#Utilize um laço for com estruturas condicionais if/elif/else para contabilizar quantos números são positivos, quantos são negativos e quantos são exatamente zero. Ao final, imprima esse resumo.

lista_numeros = [0, 2, -3, 5, 9, 8, -7, 0, 6, 1, -4]

positivo = 0
negativo = 0
zero = 0

for numero in lista_numeros:
  if numero > 0:
    positivo += 1
  elif numero < 0:
    negativo += 1
  else:
    zero += 1

print(lista_numeros)
print(f'{positivo} números da lista são positivos.')
print(f'{negativo} números da lista são negativos.')
print(f'{zero} números da lista são exatamente zero.')

```
